### Key Observations

## Missing Values

1. rating — 8,626 nulls (~31.3%) — This is the most critical gap. Over 1/3rd of products have no rating, which could heavily impact any recommendation or ranking model.
2. description — 115 nulls (~0.4%) — Minor but worth imputing or filtering for NLP/embedding-based tasks.

## No Duplicates

## Pricing Insights

1. High standard deviation relative to mean → significant price skew / long-tail distribution (many cheap items, a few very expensive ones).
2. The gap between sale and market price suggests discounting is common.

## Engineered Feature

1. Mean discount: ~11.8%
2. Median discount: 5% (50th percentile)
3. Max discount: 83.67% — some products are heavily discounted
4. 25th percentile = 0% → at least a quarter of products have no discount at all

## Ratings Distribution

1. Rated products only (18,929 records)
2. Mean: 3.94 out of 5
3. Std Dev: 0.74 — fairly concentrated
4. 25th–75th percentile: 3.7 to 4.3 — most products cluster in a "good" range
5. Only 40 unique rating values (likely increments of 0.1)

## Category Dominance

Beauty & Hygiene dominates at nearly 3× the second category — expect class imbalance if building category classifiers.


## Recommendations

1. Handle rating nulls as most of the customers didnt review the product they purchased.
2. description nulls to fill with empty string or a placeholder like "No description available"
3. Category imbalance as Beauty & Hygiene is ~28.5% of the data; using stratified sampling when splitting can help us give some accurate results.
